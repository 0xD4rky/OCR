    Align columns in the 2D array by shifting text left if the adjacent column is empty.
    
    :param out_array: 2D array of text data.
    :return: Aligned 2D array.
    """
    df = pd.DataFrame(out_array)  # Convert to DataFrame
    for i in range(1, df.shape[1]):  # Iterate through columns
        for j in range(df.shape[0]):  # Iterate through rows
            if df.iloc[j, i] and not df.iloc[j, i - 1]:  # If current cell has value but previous cell is empty
                df.iloc[j, i - 1] = df.iloc[j, i]  # Move value to previous cell
                df.iloc[j, i] = ""  # Clear current cell
    return df.values

```python
def align_columns(out_array):
  """ Align columns in the 2D array by shifting text left if the adjacent column is empty.

  :param out_array: 2D array of text data.
  :return: Aligned 2D array.
  """
  df = pd.DataFrame(out_array)  # Convert to DataFrame
  for i in range(1, df.shape[1]):  # Iterate through columns
      for j in range(df.shape[0]):  # Iterate through rows
          if df.iloc[j, i] and not df.iloc[j, i - 1]:  # If current cell has value but previous cell is empty
              df.iloc[j, i - 1] = df.iloc[j, i]  # Move value to previous cell
              df.iloc[j, i] = ""  # Clear current cell
  return df.values
```

```python
def remove_incomplete_duplicates(out_array):
  """ Remove columns with incomplete or duplicate data from the 2D array.

  :param out_array: 2D array of text data.
  :return: Cleaned 2D array.
  """
  df = pd.DataFrame(out_array)  # Convert to DataFrame
  df_cleaned = df.copy()  # Copy DataFrame for cleaning
  for col in df.columns:
      if df[col].isnull().sum() > 0:  # Check if column has any null values
          df_cleaned = df_cleaned.drop(columns=[col])  # Drop columns with null values
  return df_cleaned.values
```

```python
def ocr_to_csv(url, csv_file_path):
    """
    Perform OCR on the image and save the extracted text to a CSV file.

    :param image_path: Path to the input image file.
    :param csv_file_path: Path where the CSV file will be saved.
    """

    image = fetch_image(url)
    image_cv = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)  # Convert image to OpenCV format

    image_height, image_width = image_cv.shape[:2]  # Get image dimensions
    output = ocr.ocr(image_cv)[0]  # Perform OCR using PaddleOCR
    # boxes = [line[0] for line in output]  # Extract bounding boxes
    # texts = [line[1][0] for line in output]  # Extract text
    # probabilities = [line[1][1] for line in output]  # Extract probabilities
    if output:
      for line in output[0]:
        boxes = [line[0] for line in output]
        # Extract text, use an empty string if text is None or empty
        texts = [line[1][0] for line in output]
        probabilities = [line[1][1] for line in output]  # Extract probabilities
    else:
      # If no OCR output, consider adding default values or handling as needed
      boxes = [[[0, 0], [0, 0], [0, 0], [0, 0]]]
      texts = [""]
      probabilities = [0.0]
      # Draw bounding boxes and text on the image

    image_boxes = image_cv.copy()
    for box, text in zip(boxes, texts):
        cv2.rectangle(image_boxes, (int(box[0][0]), int(box[0][1])), (int(box[2][0]), (int(box[2][1]))), (0, 0, 255), 1)
        cv2.putText(image_boxes, text, (int(box[0][0]), int(box[0][1])), cv2.FONT_HERSHEY_SIMPLEX, 1, (222, 0, 0), 1)

    # Prepare boxes for non-maximum suppression
    horiz_boxes, vert_boxes = [], []
    for box in boxes:
        x_h, x_v = 0, int(box[0][0])
        y_h, y_v = int(box[0][1]), 0
        width_h, width_v = image_width, int(box[2][0] - box[0][0])
        height_h, height_v = int(box[2][1] - box[0][1]), image_height
        horiz_boxes.append([x_h, y_h, x_h + width_h, y_h + height_h])
        vert_boxes.append([x_v, y_v, x_v + width_v, y_v + height_v])

    # Apply non-maximum suppression to remove redundant boxes
    horiz_out = tf.image.non_max_suppression(
        horiz_boxes,
        probabilities,
        max_output_size=1000,
        iou_threshold=0.1,
        score_threshold=float('-inf')
    )
    horiz_lines = np.sort(np.array(horiz_out))

    vert_out = tf.image.non_max_suppression(
        vert_boxes,
        probabilities,
        max_output_size=1000,
        iou_threshold=0.1,
        score_threshold=float('-inf')
    )
    vert_lines = np.sort(np.array(vert_out))

    # Create a 2D array to store text aligned by rows and columns
    out_array = [["" for _ in range(len(vert_lines))] for _ in range(len(horiz_lines))]
    unordered_boxes = [vert_boxes[i][0] for i in vert_lines]
    ordered_boxes = np.argsort(unordered_boxes)

    def intersection(box_1, box_2):
        """Compute intersection of two bounding boxes."""
        return [box_2[0], box_1[1], box_2[2], box_1[3]]

    def iou(box_1, box_2):
        """Calculate Intersection over Union (IoU) of two bounding boxes."""
        x_1 = max(box_1[0], box_2[0])
        y_1 = max(box_1[1], box_2[1])
        x_2 = min(box_1[2], box_2[2])
        y_2 = min(box_1[3], box_2[3])
        inter = abs(max((x_2 - x_1, 0)) * max((y_2 - y_1, 0)))
        if inter == 0:
            return 0
        box_1_area = abs((box_1[2] - box_1[0]) * (box_1[3] - box_1[1]))
        box_2_area = abs((box_2[2] - box_2[0]) * (box_2[3] - box_2[1]))
        return inter / float(box_1_area + box_2_area - inter)

    # Fill in the out_array with detected text
    for i in range(len(horiz_lines)):
        for j in range(len(vert_lines)):
            resultant = intersection(horiz_boxes[horiz_lines[i]], vert_boxes[vert_lines[ordered_boxes[j]]])
            for b in range(len(boxes)):
                the_box = [boxes[b][0][0], boxes[b][0][1], boxes[b][2][0], boxes[b][2][1]]
                if iou(resultant, the_box) > 0.1:
                    out_array[i][j] = texts[b]

    out_array = np.array(out_array)

    # Post-process the array to align columns correctly and remove incomplete duplicates
    out_array = align_columns(out_array)
    out_array = remove_incomplete_duplicates(out_array)

    # Save the processed data to a CSV file
    with open(csv_file_path, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerows(out_array)
    print(f"Data successfully saved to {csv_file_path}")
```

```python
def fetch_image(url):
    """
    Fetch image from the provided URL.
    """
    response = requests.get(url)
    image = cv2.imdecode(np.frombuffer(response.content, np.uint8), cv2.IMREAD_COLOR)
    if image is None:
        raise ValueError("Failed to decode the image from the provided URL.")
    return image
```

```python
import re
import pandas as pd
import numpy as np

# Given entity_unit_map
entity_unit_map = {
    'width': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'depth': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'height': {'centimetre', 'foot', 'inch', 'metre', 'millimetre', 'yard'},
    'item_weight': {'gram', 'kilogram', 'microgram', 'milligram', 'ounce', 'pound', 'ton'},
    'maximum_weight_recommendation': {'gram', 'kilogram', 'microgram', 'milligram', 'ounce', 'pound', 'ton'},
    'voltage': {'kilovolt', 'millivolt', 'volt'},
    'wattage': {'kilowatt', 'watt'},
    'item_volume': {'centilitre', 'cubic foot', 'cubic inch', 'cup', 'decilitre', 'fluid ounce',
                    'gallon', 'imperial gallon', 'litre', 'microlitre', 'millilitre', 'pint', 'quart', 'ounce'}
}

# Function to generate unit variations
def generate_unit_variations(unit):
    variations = set()
    # Add the unit itself
    variations.add(unit)
    # Add plural form
    if not unit.endswith('s'):
        variations.add(unit + 's')
    # Common abbreviations and variations
    abbreviations = {
        'gram': ['g', 'gm', 'gr'],
        'kilogram': ['kg', 'kgs', 'kilo'],
        'milligram': ['mg', 'mgs'],
        'microgram': ['mcg'],
        'ounce': ['oz', 'fl oz'],
        'pound': ['lb', 'lbs'],
        'ton': ['t'],
        'metre': ['m', 'meter', 'meters'],
        'centimetre': ['cm', 'centimeter', 'centimeters'],
        'millimetre': ['mm', 'millimeter', 'millimeters'],
        'inch': ['in', 'inches', '"'],
        'foot': ['ft', "'"],
        'yard': ['yd', 'yds'],
        'volt': ['v', 'volts'],
        'kilovolt': ['kv', 'kilovolts'],
        'millivolt': ['mv', 'millivolts'],
        'watt': ['w', 'watts'],
        'kilowatt': ['kw', 'kilowatts'],
        'litre': ['l', 'liter', 'liters'],
        'millilitre': ['ml', 'milliliter', 'milliliters'],
        'microlitre': ['µl', 'microliter', 'microliters'],
        'gallon': ['gal', 'gals'],
        'pint': ['pt', 'pts'],
        'quart': ['qt', 'qts'],
        'cup': ['cups'],
        'fluid ounce': ['fl oz', 'fluid oz'],
        'cubic foot': ['ft³', 'cubic ft'],
        'cubic inch': ['in³', 'cubic in'],
        'centilitre': ['cl', 'centiliter', 'centiliters'],
        'decilitre': ['dl', 'deciliter', 'deciliters'],
        'imperial gallon': ['imp gal', 'imperial gal'],
        'ounce': ['oz', 'oz.', 'ounce', 'ounces'],
    }
    if unit in abbreviations:
        variations.update(abbreviations[unit])
    # Handle capitalization variations
    variations.update({v.upper() for v in variations})
    variations.update({v.capitalize() for v in variations})
    return variations

# Function to create regex patterns for each entity
def create_entity_patterns(entity_unit_map):
    entity_patterns = {}
    for entity, units in entity_unit_map.items():
        unit_variations = set()
        for unit in units:
            unit_variations.update(generate_unit_variations(unit))
        # Create regex pattern for units
        unit_pattern = r'|'.join(sorted(map(re.escape, unit_variations), key=len, reverse=True))
        # Regex pattern to find numbers and units in various formats
        pattern = re.compile(
            rf'(?i)(?P<number>\b\d+(?:[\.,]\d+)?|\b(?:{"|".join(NUMBER_WORDS.keys())})(?:\s+(?:{"|".join(NUMBER_WORDS.keys())}))*\b)[\s\-]*'
            rf'(?P<unit>{unit_pattern})\b|'
            rf'(?P<unit_before>{unit_pattern})[\s\-]*(?P<number_before>\b\d+(?:[\.,]\d+)?|\b(?:{"|".join(NUMBER_WORDS.keys())})(?:\s+(?:{"|".join(NUMBER_WORDS.keys())}))*\b)'
        )
        entity_patterns[entity] = pattern
    return entity_patterns

# Number words mapping
NUMBER_WORDS = {
    'zero': 0,
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven':7,
    'eight':8,
    'nine':9,
    'ten':10,
    'eleven':11,
    'twelve':12,
    'thirteen':13,
    'fourteen':14,
    'fifteen':15,
    'sixteen':16,
    'seventeen':17,
    'eighteen':18,
    'nineteen':19,
    'twenty':20,
    'thirty':30,
    'forty':40,
    'fifty':50,
    'sixty':60,
    'seventy':70,
    'eighty':80,
    'ninety':90,
    'hundred':100,
    'thousand':1000,
    'million':1000000,
    'billion':1000000000
}

# Function to convert number words to numeric value
def text2int(textnum):
    tokens = re.findall(r'\w+', textnum.lower())
    current = result = 0
    for word in tokens:
        if word not in NUMBER_WORDS:
            continue
        scale = NUMBER_WORDS[word]
        if scale >= 100:
            current *= scale
        else:
            current += scale
    return str(current + result)

# Function to extract entity values from text
def extract_entity_values(row, patterns):
    text = row['text']
    entity = row['entity_name']
    if pd.isna(text) or text.strip() == '':
        return ''
    pattern = patterns.get(entity)
    if not pattern:
        return ''
    matches = pattern.finditer(text)
    for match in matches:
        if match.group('number') and match.group('unit'):
            number = match.group('number')
            unit = match.group('unit')
        elif match.group('number_before') and match.group('unit_before'):
            number = match.group('number_before')
            unit = match.group('unit_before')
        else:
            continue
        # Convert number words to digits
        if not re.match(r'\d', number):
            number = text2int(number)
        # Replace comma with dot for decimals
        number = number.replace(',', '.')
        return f"{number} {unit}"
    return ''


df = pd.read_csv('/content/extracted_2.csv')

# Create patterns
patterns = create_entity_patterns(entity_unit_map)

# Apply extraction function
df['extracted_value'] = df.apply(extract_entity_values, axis=1, patterns=patterns)

```
